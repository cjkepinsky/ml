```python
from libs.simple_processing import separate_target, scale
import pandas as pd
from constants import preprocessed_data

df = pd.read_csv(preprocessed_data)

scaled = scale(df)

length = int(len(scaled) * 0.99)
TEST = scaled[length:]
DATA = scaled[:length]

DATA
```

```python

from constants import target_name, x_path, y_path, x_test_path, y_test_path

X, y = separate_target(DATA, target_name)
Xt, yt = separate_target(TEST, target_name)

X.to_csv(x_path, index=False)
y.to_csv(y_path, index=False)
Xt.to_csv(x_test_path, index=False)
yt.to_csv(y_test_path, index=False)

Xt
```
