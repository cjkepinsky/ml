```python
from libs.simple_analyzer import print_analytics, print_simple_correlations
from libs.simpleplotter import simple_heatmap, simple_correlations
from constants import raw_data, target_name
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

DATA = pd.read_csv(raw_data)

print(DATA.isna().sum())
```

```python
# print_analytics(DATA, target_name)
# Getting the main parameters of the Normal Ditribution ()
# https://www.kaggle.com/code/marto24/beginners-prediction-top3
import matplotlib.pyplot as plt
from scipy.stats import skew, norm
import seaborn as sns


(mu, sigma) = norm.fit(DATA['SalePrice'])

plt.figure(figsize = (12,6))
sns.distplot(DATA['SalePrice'], kde = True, hist=True, fit = norm)
plt.title('SalePrice distribution vs Normal Distribution', fontsize = 13)
plt.xlabel("House's sale Price in $", fontsize = 12)
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
           loc='best')
plt.show()
```
